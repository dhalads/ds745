```python
import networkx as nx
from random import *
import matplotlib as plt
```

Create a complete graph where all nodes connect to all other nodes:

```python
n = 20
G = nx.complete_graph(n)

print(G.nodes)

print(G.edges)
```

```python
nx.draw(G)
```

Delete some edges because in the real world we won't have all nodes connecting to all nodes:

```python
for i in range(len(G.edges)-1,0,-1):
    if i % randint(1,3) == 0:
        e=list(G.edges())[i]
        G.remove_edge(*e)
```

```python
nx.draw(G)
```

Pick some arbitrary bad guy nodes:

```python
bad_guy_nodes = [0,3,8]
```

Add an attribute to every node, indicating whether a node is a good guy or bad guy.

```python
for i in range(0, len(G.nodes())):
    if i in bad_guy_nodes:
        G.nodes[list(G.nodes())[i]]['good_guy'] = 0
    else:
        G.nodes[list(G.nodes())[i]]['good_guy'] = 1

print(G.nodes.data())
```

This is how you get a list of attribute values:

```python
list(nx.get_node_attributes(G,'good_guy').values())
```

Draw the graph, highlighting the bad guy nodes in a different color than the good guy nodes:

```python
nx.draw(G, node_color = list(nx.get_node_attributes(G,'good_guy').values()), vmin=0, vmax=1, cmap = plt.cm.get_cmap('autumn'))
```

Filter the nodes to ONLY retrieve the ones that are NOT bad guys

```python
good_guy_nodes = [x for x,y in G.nodes(data=True) if y['good_guy'] == 1]
print(good_guy_nodes)
```
